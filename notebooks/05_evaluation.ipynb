## Note — the threshold is a bumpy dial, not a free choice

The decision threshold trades two costs against each other, and both extremes are bad:

- **Threshold → 0:** almost every score clears the bar, so the model flags *everything* as fraud. It blocks nearly every transaction. Fraud recall is ~perfect, but legitimate customers get declined — friction, lost revenue, churn.
- **Threshold → 1:** almost nothing clears the bar — a "free bird" that waves everything through. Real fraud sails past unblocked, so it costs more money in direct fraud losses.

So the right threshold sits somewhere in between, and *where* depends on the relative £ cost of a false positive (a blocked legit transaction) vs a false negative (a missed fraud). That's why we tune it against a cost curve in £ rather than picking the threshold that maximises F1 — F1 treats both errors as equal, but the business doesn't.
